### UNIFYING DATAFRAMES FROM DIFFERENT SOURCES

concat clean df from inmuebles24.com and lamundi.com which contain different data from real state adds
we have droped several columns due to unmatching variables
the selected columns are the minimum necessary to run a model


```python
import pandas as pd
import numpy as np
```

```python
inmb24_dat = pd.read_csv('../00_scrap_inmuebles24/00_inmuebles24_data_processed.csv', sep='|')

inmb24_dat['categoria'] = 'departamento'
inmb24_dat= inmb24_dat[['precio_m2',  'superficie_total', 'categoria',
                        'dormitorios', 'baños', 'cocheras',
                       'colonia', 'alcaldia', 'lat', 'lng']]
inmb24_dat = inmb24_dat.dropna()

inmb24_dat
```

```python
lamundi_dat = pd.read_csv('../00_scrap_lamundi/00_Lamudi_houses_sale.csv')
lamundi_dat = lamundi_dat.reset_index(drop=True).iloc[:,1:]
lamundi_dat['price']= pd.to_numeric(lamundi_dat['price'], errors='coerce')
lamundi_dat['land_size']= pd.to_numeric(lamundi_dat['land_size'], errors='coerce')

lamundi_dat['precio_m2'] = lamundi_dat['price'] / lamundi_dat['land_size']

lamundi_dat= lamundi_dat.rename(columns={'category':'categoria',
                                         'land_size' :'superficie_total', 
                                        'car_spaces': 'cocheras', 
                                        'bedrooms': 'dormitorios', 
                                        'bathrooms': 'baños'})

lamundi_dat['cocheras'] = pd.to_numeric(lamundi_dat['cocheras'], errors='coerce')
lamundi_dat['dormitorios']= pd.to_numeric(lamundi_dat['dormitorios'], errors='coerce')
lamundi_dat['baños']= pd.to_numeric(lamundi_dat['baños'], errors='coerce')

lamundi_dat['colonia'] = np.nan
lamundi_dat['alcaldia'] = np.nan
lamundi_dat['lng'] = [pd.to_numeric(lamundi_dat['location'][i].split(',')[0].replace('[', '')) for i in range(0, len(lamundi_dat))]
lamundi_dat['lat'] = [pd.to_numeric(lamundi_dat['location'][i].split(',')[1].replace(']', '')) for i in range(0, len(lamundi_dat))]
lamundi_dat = lamundi_dat[['precio_m2',  'superficie_total', 'categoria',
                           'dormitorios', 'baños', 'cocheras',
                           'colonia', 'alcaldia', 'lat', 'lng']]
lamundi_dat.count()
```

```python
lamundi_dat2 = pd.read_csv('../00_scrap_lamundi/00_lamudi_departments_sale.csv')

lamundi_dat2 = lamundi_dat2.reset_index(drop=True).iloc[:,1:]
lamundi_dat2['price']= pd.to_numeric(lamundi_dat2['price'], errors='coerce')
lamundi_dat2['land_size']= pd.to_numeric(lamundi_dat2['building_size'], errors='coerce')

lamundi_dat2['precio_m2'] = lamundi_dat2['price'] / lamundi_dat2['land_size']

lamundi_dat2= lamundi_dat2.rename(columns={'category':'categoria',
                                         'land_size' :'superficie_total', 
                                        'car_spaces': 'cocheras', 
                                        'bedrooms': 'dormitorios', 
                                        'bathrooms': 'baños'})

lamundi_dat2['cocheras'] = pd.to_numeric(lamundi_dat2['cocheras'], errors='coerce')
lamundi_dat2['dormitorios']= pd.to_numeric(lamundi_dat2['dormitorios'], errors='coerce')
lamundi_dat2['baños']= pd.to_numeric(lamundi_dat2['baños'], errors='coerce')

lamundi_dat2['colonia'] = np.nan
lamundi_dat2['alcaldia'] = np.nan
lamundi_dat2['lng'] = [pd.to_numeric(lamundi_dat2['coordinates'][i].split(',')[0].replace('[', '')) for i in range(0, len(lamundi_dat2))]
lamundi_dat2['lat'] = [pd.to_numeric(lamundi_dat2['coordinates'][i].split(',')[1].replace(']', '')) for i in range(0, len(lamundi_dat2))]
lamundi_dat2 = lamundi_dat2[['precio_m2',  'superficie_total', 'categoria',
                           'dormitorios', 'baños', 'cocheras',
                           'colonia', 'alcaldia', 'lat', 'lng']]
lamundi_dat2.count()
```

```python
dat = inmb24_dat.append(lamundi_dat, ignore_index=True)
dat
```

```python
data_total = dat.append(lamundi_dat2, ignore_index=True)
data_total.count()
```

```python
data_total.to_csv('01_data_total_unified.csv')
```
