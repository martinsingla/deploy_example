## Scrapper for: "inmuebles24.com"

```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import time
```

## GETTING URLS OF PUBLISHED ADDS

```python
crawled_urls = []

########################################
import warnings
warnings.filterwarnings('ignore')

print('Crawling base url')

executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
url_init = 'https://www.inmuebles24.com/casas-o-departamentos-en-venta-en-distrito-federal'

browser.visit(url_init + '.html')
html = browser.html
soup = bs(html, 'html.parser')
browser.quit()

cards = soup.find_all('div', class_= 'postingCardInfo')
for i in range(0, len(cards)):
    try:
        href = 'https://www.inmuebles24.com' + cards[i].find_all('a', 'go-to-posting')[0]['href']
        crawled_urls.append(href)
    except:
        pass
    
print(f'Got: {len(crawled_urls)} urls')
time.sleep(5)

for i in range(1, 200):
    print('Crawling url', i)
    
    executable_path = {'executable_path': ChromeDriverManager().install()}
    browser = Browser('chrome', **executable_path, headless=False)
    browser.visit(url_init + '-pagina-' + str(i) +'.html')
    html = browser.html
    soup = bs(html, 'html.parser')
    browser.quit()
    
    try:
        cards = soup.find_all('div', class_= 'postingCardInfo')
        for i in range(0, len(cards)):
            try:
                href = 'https://www.inmuebles24.com' + cards[i].find_all('a', 'go-to-posting')[0]['href']
                crawled_urls.append(href)
            except:
                pass
            
    except:
        pass
    
    time.sleep(5)
    print(f'Got: {len(crawled_urls)} urls now!!')

########################################
df = pd.DataFrame({
    'urls': crawled_urls
})
df.to_csv('scrapped_urls.csv')
```

## GETTING ALL DATA FROM INDIVIDUAL ADDS

```python
urls = pd.read_csv('scrapped_urls.csv')
urls= urls['urls']
```

```python
adds = {}

for i in range(0, len(urls)):
    
    executable_path = {'executable_path': ChromeDriverManager().install()}
    browser = Browser('chrome', **executable_path, headless=False)
    
    time.sleep(2)
    browser.visit(urls[i])

    html = browser.html
    soup = bs(html, 'html.parser')
    
    browser.quit()
    
    add_dict = {
        'id': [i],
        'url': [urls[i]],
        'operacion': [],
        'direccion': [],
        'antiguedad': [],
        'descripcion': [],
        'superficie_total': [],
        'superficie_cubierta': [],
        'baos': [],
        'dormitorios': [],
        'cocheras': [],
        'precio': [],
        'mantenimiento': [],
        'google_src': [],
        'lat': [],
        'lng': [],
        'varios': [],
        'anunciante': []
    }
    
    print('Attempting to scrap data from:', i)
    
    try:
        try:
            #operacion
            add_dict['operacion'] = soup.find_all('div', class_='price-operation')[0].text
        except:
            print('Not able to get "operacin" ')

        try:
            #direccion
            add_dict['direccion'] = soup.find_all('h2', class_= 'title-location')[0].text
        except:
            try:
                add_dict['direccion'] = soup.find_all('div', class_='section-location')[0].text
            except:
                print('Not able to get "direccion" ')

        try:
            #descripcin
            add_dict['descripcion'] = soup.find_all('div', id= 'longDescription')[0].text
        except:
            print('Not able to get "descripcion" ')

        try:
            a= soup.find_all('li', class_='icon-feature')
        except:
            print('Not able to get "icon feature list" ')
        
        for j in range(0, len(a)):
            
            try:
                #Antiguedad
                if len(a[j].find_all(class_='icon-antiguedad')) > 0:
                    add_dict['antiguedad'] = a[j].text.replace('\n', '').replace('\t', '') 
            except:
                print('Not able to get "Antiguedad"')
            
            try:
                #Superficie total:
                if len(a[j].find_all(class_='icon-stotal')) > 0:
                    add_dict['superficie_total'] = a[j].text.replace('\n', '').replace('\t', '')
            except:
                print('Not able to get "Superficie total" ')

            try:
                #Superficie cubierta:
                if len(a[j].find_all(class_= 'icon-scubierta')) > 0:
                    add_dict['superficie_cubierta'] = a[j].text.replace('\n', '').replace('\t', '')
            except:
                print('Not able to get "Superficie cubierta" ')
    
            try:
                #Baos
                if len(a[j].find_all(class_= 'icon-bano')) > 0:
                    add_dict['baos'] = a[j].text.replace('\n', '').replace('\t', '')
            except:
                print('Not able to get "Superficie baos" ')

            try:
                #Cocheras
                if len(a[j].find_all(class_= 'icon-cochera')) > 0:
                    add_dict['cocheras'] = a[j].text.replace('\n', '').replace('\t', '')
            except:
                print('Not able to get "cocheras" ')

            try:
                #Dormitorios
                if len(a[j].find_all(class_= 'icon-dormitorio')) > 0:
                    add_dict['dormitorios'] = a[j].text.replace('\n', '').replace('\t', '') 
            except:
                print('Not able to get "dormitorios" ')

        try:
            #precio publicado
            add_dict['precio'] = soup.find_all('div', class_='price-items')[0].find_all('span')[0].text.replace('\n', '').replace('\t', '')
        except:
                print('Not able to get "precio" ')
        
        try:
            #precio mantenimiento
            add_dict['mantenimiento'] = soup.find_all('div', class_= 'block-expensas block-row')[0].find_all('span')[0].text
        except:
            print('Not able to get "mantenimiento"')

        try:
            #Google SRC map
            add_dict['google_src'] =soup.find_all('img', id= 'static-map')[0]['src']
        except:
                print('Not able to get "map" ')

        try:
            #Coordenadas
            add_dict['lat'] = add_dict['google_src'].split('&markers=')[1].split('&')[0].split(',')[0]
            add_dict['lng'] = add_dict['google_src'].split('&markers=')[1].split('&')[0].split(',')[1]
        except:
            print('Not able to get "Coordenadas" ')
                
        try:
            #anunciante
            add_dict['anunciante'] = soup.find_all('h3', class_= 'sc-dPaNzc gdkhio')[0].text
        except:
            print('Not able to get "anunciante"')

            
        try:
            #Caracteristicas generales
            ul = soup.find_all('ul', class_= 'sc-bYwzuL fDDjHD')[0].find_all('li')
            varios = []
            [varios.append(l.text) for l in ul]
            varios
            add_dict['varios'] = varios
        except:
            print('Not able to get "caracteristicas generales"')

        adds[i] = add_dict
        
    except:
        print('not able to get data')
        pass


```

```python
import json

json_object = json.dumps(adds)

with open("scrapped_adds.json", "w") as outfile:
    outfile.write(json_object)
```

```python
df_adds = pd.DataFrame.from_dict(adds, orient='index')

for i in df_adds.columns:
    df_adds[i].str.replace('|', '')
    
df_adds.to_csv('scrapped_adds.csv', sep= '|')    
```
