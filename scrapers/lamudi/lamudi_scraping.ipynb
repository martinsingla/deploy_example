```python
# Dependencies
from splinter import Browser
from bs4 import BeautifulSoup as bs
from webdriver_manager.chrome import ChromeDriverManager
import time
import math
import pandas as pd
```

```python
# Number of pages to visit
page = 100
counter = 0 
```

```python

# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python

#URL_rent
# https://www.lamudi.com.mx/distrito-federal/casa/for-sale/?currency=mxn&page
# https://www.lamudi.com.mx/distrito-federal/departamento/for-sale/?currency=mxn&page
# https://www.lamudi.com.mx/distrito-federal/departamento/for-rent/?currency=mxn&page
# https://www.lamudi.com.mx/distrito-federal/casa/for-rent/?currency=mxn&page 
```

```python
list_bathrooms=[]
list_bedrooms=[]
list_rooms_total=[]
list_building_size=[]
list_car_spaces=[]
list_category=[]
list_furnished=[]
list_coordinates=[]
list_location=[]
list_land_size=[]
list_price=[]
list_id=[]
list_subcategory=[]
list_year_built=[]
list_description=[]
list_link=[]
```

```python
for i in range(0,page):
    counter = counter + 1 
    counter_str = str(counter)
    url = 'https://www.lamudi.com.mx/distrito-federal/departamento/for-sale/?currency=mxn&page=' + counter_str
    browser.visit(url)
    time.sleep(1)
    html = browser.html
    soup = bs(html, 'html.parser')
    
    des=[]
    features = soup.find_all('div', class_="ListingCell-AllInfo ListingUnit")
    other_features = soup.find_all('a', class_= 'js-listing-link ellipsis')
    
    for i in features:
        extra_features=soup.find_all('div', class_="ListingCell-shortDescription")
        for i in range(len(extra_features)):
            d=extra_features[i].get_text(strip = True).replace("\n\n", "").replace("\n", " ").replace("\t", "").replace("\xa0", "")
            des.append(d) 
    loc=[]        
    location=soup.find_all('span', class_="ListingCell-KeyInfo-address-text")
    for i in range(len(location)):
        l=location[i].get_text(strip = True).replace("\n\n", "").replace("\n", " ")
        loc.append(l) 
        
    for i in range(0,30):
      
        try:
            bathrooms=features[i]['data-bathrooms']
            list_bathrooms.append(bathrooms)
        except:
            bathrooms="uknown"
            list_bathrooms.append(bathrooms)  
            
        try:    
            bedrooms=features[i]['data-bedrooms']
            list_bedrooms.append(bedrooms)       
        except:
            bedrooms="unknown"
            list_bedrooms.append(bedrooms)
                
        try:
            rooms_total=features[i]['data-rooms_total']
            list_rooms_total.append(rooms_total)
        except:
            rooms_total="unknown"
            list_rooms_total.append(rooms_total)

            
        try:    
            building_size=features[i]['data-building_size']
            list_building_size.append(building_size)
        except:
            building_size="unknown"
            list_building_size.append(building_size)          

            
        try:
            car_spaces=features[i]['data-car_spaces']
            list_car_spaces.append(car_spaces)   
        except:
            car_spaces="unknown"
            list_car_spaces.append(car_spaces)

            
        try:
            category=features[i]['data-category']
            list_category.append(category)
        except:
            category="unknown"
            list_category.append(category)

            
        try:
            furnished=features[i]['data-furnished']
            list_furnished.append(furnished)
        except:
            furnished="unknown"
            list_furnished.append(furnished)

            
        try:
            coordinates=features[i]['data-geo-point']
            list_coordinates.append(coordinates)
        except:
            coordinates="unknown"
            list_coordinates.append(coordinates)
            
        try:
            location=loc[i]
            list_location.append(location)
        except:
            location="unknown"
            list_location.append(location)    

            
        try:
            land_size=features[i]['data-land_size']
            list_land_size.append(land_size)
        except:
            land_size="unknown"
            list_land_size.append(land_size)

            
        try:
            price=features[i]['data-price']
            list_price.append(price)
        except:
            price="unknown"
            list_price.append(price)

            
        try:
            id=features[i]['data-sku']
            list_id.append(id)
        except:
            id="unknown"
            list_id.append(id)

            
        try:
            subcategory=features[i]['data-subcategories']
            list_subcategory.append(subcategory)
        except:
            subcategory="unknown"
            list_subcategory.append(subcategory)

            
        try:
            year_built=features[i]['data-year_built']
            list_year_built.append(year_built)
        except:
            year_built="unknown"
            list_year_built.append(year_built)

            
        try:
            description=des[i]
            list_description.append( description)
        except:
            description="unknown"
            list_description.append( description)

            
        try:
            link=other_features[i]['href']
            list_link.append(link)
        except:
            link="unknown"
            list_link.append(link)   



```

```python
dict = {"bathrooms":list_bathrooms,
        "bedrooms":list_bedrooms,
        "rooms_total":list_rooms_total,
        "building_size":list_building_size,
        "car_spaces":list_car_spaces,
        "category":list_category,
        "furnished":list_furnished,
        "coordinates":list_coordinates,
        "location":list_location,
        "land_size":list_land_size,
        "price":list_price,
        "id":list_id,
        "subcategory":list_subcategory,
        "year_built": list_year_built,
        "description":list_description,
        "link":list_link
        }
        
```

```python
df = pd.DataFrame(dict)
```
