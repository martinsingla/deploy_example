```python
# Libraries and dependencies
import json
import pandas as pd
import unidecode
import csv
```

```python
# Opening the json
with open('viva_anuncios.json') as json_file:
    raw_dict = json.load(json_file)
    json_file.close()
        
```

```python
raw_dict
```

```python
# Opening the csv
colony = pd.read_csv('coloniascdmx.csv')

colony
```

```python
# Keeping relevant columns
colony = colony[['nombre', 'geo_point_2d']]
colony[['latitude', 'longitude']] = colony.geo_point_2d.str.split(',', expand=True)
colony.drop(columns=['geo_point_2d'], inplace=True)

colony
```

```python
colony['nombre']
```

```python
# Casting adress and description to all uppercases
for ad in raw_dict:
    raw_dict[ad]['address'] = raw_dict[ad]['address'].upper()
    raw_dict[ad]['description'] = raw_dict[ad]['description'].upper()

# Getting rid of the accents
for ad in raw_dict:
    raw_dict[ad]['address'] = unidecode.unidecode(raw_dict[ad]['address'].upper())
    raw_dict[ad]['description'] = unidecode.unidecode(raw_dict[ad]['description'].upper())

raw_dict   
```

```python
# Appending colony

for ad in raw_dict:
    for row in range(0,len(colony)):
        sub_str = colony['nombre'][row]
        str1 = raw_dict[ad]['address']
        str2 = raw_dict[ad]['description']
        if str1.find(sub_str) != -1 or str2.find(sub_str) != -1:
            raw_dict[ad]['colony'] = sub_str
            raw_dict[ad]['latitude']= colony['latitude'][row]
            raw_dict[ad]['longitude']= colony['latitude'][row]
```

```python
raw_dict
```

```python
# Saving the processed data to a json
with open('viva_anuncios_processed.json', 'w') as fp:
    json.dump(raw_dict, fp, indent=4)
    fp.close()
```

```python
# Cleaning
data =pd.DataFrame(raw_dict)
data = data.transpose()

data
```

```python
# Cleaning square meters
square_meters = data['square_meters']

for _ in range(0, len(square_meters)):
    try:
        square_meters[_] = int(square_meters[_])
    except:
        square_meters[_] = 0
data['square_meters'] = square_meters
data = data.astype({'square_meters': 'int32', 'price': 'int32'})
data.dtypes
```
